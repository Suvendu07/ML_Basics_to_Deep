```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
# Load the data into DataFrame
df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
df.head()
```

```python
# Add the target columns
df['label'] = breast_cancer.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
# split the dataset
x = df.drop(columns='label', axis=1)
y = df['label']
```

```python
x
```

```python
y
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

```python
x
```

```python
y
```

```python
# GridSearchCV
model = SVC()
```

```python
# Hyperparameters
parameters = {'kernel': ['linear', 'poly', 'rbf','sigmoid'],
              'C' : [1,5,10,20]}
```

```python
# grid search
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy
highest_accu = classifier.best_score_
print(highest_accu)
```

```python
# Loading the results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

## RandomizedSearchCV

```python
# load the SVC model
model_1 = SVC()
```

```python
# hyperparameters
parameters_1 = {
    'kernel': ['linear','poly','rbf','sigmoid'],
    'C': [1,5,10,20]
}
```

```python
# RandomSearch
classifier_1 = RandomizedSearchCV(model_1, parameters_1)
```

```python
classifier_1.fit(x,y)
```

```python
RandomizedSearchCV(cv = 5, estimator=SVC(),
                   param_distributions={'C': [1,5,10,20],
                                        'kernel': ['linear', 'poly', 'rbf','sigmoid']})
```

```python
classifier_1.cv_results_
```

```python
# best parameters
best_parameters_1 =classifier_1.best_params_
```
